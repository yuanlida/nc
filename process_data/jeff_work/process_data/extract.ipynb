```python
from tqdm import tqdm
# import regex as re
from collections import OrderedDict
import nltk
import re
in_dir = "/Users/macos/Desktop/Edison-ai/sig-extract/enron_label/body_two.txt"
out_dir = "/Users/macos/thunder/ner-tf/process_data/enron_sig.txt"

# extracting single text file
doc = []
n = 0
lines_seen = set()
with open(in_dir, 'r') as r, open(out_dir, 'w') as w:
    for line in r.readlines():
        if line.startswith('#sig#'):
            line = line.replace('#sig#', '')
            n += 1 
            # line = line + "\n"
            if line not in lines_seen:# and line != '\n':
                lines_seen.add(line)
                line_split = re.findall(r"\w+|[^\w\s]", line, re.UNICODE)
                for word in line_split:
                    w.writelines(word)
                    w.writelines('\n')
                w.writelines('\n')
    print(n)

print('ok')
```

```python
file_dir = "/Users/macos/Desktop/Edison-ai/sig-extract/enron_test2/P/"
out_dir = "/Users/macos/thunder/ner-tf/process_data/enron_sig2.txt"

# extracting multiple text files
doc = []
lines_seen = set()
for i in range(5000):
    try:
        with open((file_dir + str(i) + "_body"), 'r') as r:
            for line in r.readlines():
                if line.startswith('#sig#'):
                    line = line.replace('#sig#', '')
                    doc.append(line)
    except:
        print(i, "does not exist")
        pass
with open(out_dir, 'w') as w:
    for ele in doc:
        if ele not in lines_seen:
            lines_seen.add(ele)
            w.writelines(ele)
            w.writelines('\n')
print('ok')
```
